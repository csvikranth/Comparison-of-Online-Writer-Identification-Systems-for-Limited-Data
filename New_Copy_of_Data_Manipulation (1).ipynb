```python
def gauss_prob(xt,mi,vi):

  (D,d)=xt.shape
  p=(1/((2*math.pi)**(D/2)*(np.linalg.det(vi))**(0.5)))*math.exp(-0.5*np.dot(np.dot(np.transpose(np.subtract(xt,mi)),np.linalg.inv(vi)),(np.subtract(xt,mi))))

  return(p)
```

```python
def MAPaDapt(features_path,  mean1, cov_new, weights, Tau, parameter): # only mean adaptation implemented
    #Douglas A. Reynolds, Thomas F. Quatieri, and Robert B. Dunn, "Speaker Verification Using Adapted Gaussian Mixture Models",Digital Signal Processing 10, 19-41 (2000)
    
    for iter in range(0, 20):

          #initialize
          ixx=(np.size(weights))
          N_tot=np.zeros(ixx)
 
          ixx=(np.size(mean1, axis=0), np.size(weights))
          F_tot, S_tot = np.zeros(ixx), np.zeros(ixx)
          frm_tot, llk_tot = 0, 0.0 #for average llk
          

          a=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]];

          for writer in writers[:1]:
            print(writer)
            with open(os.path.join(features_path,writer)) as file:
              csvreader=csv.reader(file)
              c=0
              for row in csvreader:
                a[c].extend(row)
                c+=1
          a=np.transpose(a)

          N, F, S, llk, n_frames =Collect_Stats(a,  mean1, cov_new, weights) #  for statistics
          print ("MAP %d ...%s, frame %d  llh %f") %(iter, writer, n_frames, sum(llk)/len(llk))
                
          if (np.size(N, axis=0) == np.size(weights) ) and (np.size(F, axis=0) == np.size(mean1, axis=0)):
            N_tot, F_tot, S_tot = N_tot +  N, F_tot +  F, S_tot +  S
            llk_tot, frm_tot = llk_tot + (sum(llk)/len(llk)), frm_tot + n_frames
          else:
            continue

           #all stat estimated above
          

          if parameter == 'Mean':
                alpha, mean_ml = N_tot/(N_tot + Tau), F_tot/N_tot # ml estimate 'mean'
                mu1, mu2=np.multiply(mean1, (1-alpha)), np.multiply(mean_ml, alpha)   
                m_new=mu1 + mu2

          else:
             print ("code only support - mean adaptation!")
             print ("config parameter has to be 'm' -program terminated!")
             exit()

          # replace the old UBM parameters with re-estimated one
          mean1=m_new
 
    return mean1, cov_new, weights


def Collect_Stats(a,  mean1, cov_new, weights):
    #estimate zero, first, second order statistics 
    
    try:
       n_mixtures=np.size(weights)  #512x1
    except:
       n_mixtures=len(weights)
    
    try:
       dim=np.size(mean1, axis=0) #57x512
    except:
       dim=len(mean1)
    
    if dim != np.size(a, axis=0):
       print ("Feature dimension mismatch with model!!!")
       exit() 


    # removing "nan or inf" frames from data
    idx=np.argwhere(np.isnan(a))
    a = np.delete(a, np.unique(idx[:,0]), axis=0) #axis=0 i.e. indicate rows, 1 - column
   
    idx=np.argwhere(np.isinf(a)) #inf
    data = np.delete(a, np.unique(idx[:,0]), axis=0)
  
    n_frames=np.size(data, axis=1)  # 57 X frames

    if n_frames == 0:
       print ("avoiding empty file after remove Nan/Inf")
       n, f, s, llk = [], [], [], []

    else:   

       #compute the GMM posteriors for the given data
       gammas, llk =GaussPosterior(a, mean1, cov_new, weights, n_mixtures, dim, n_frames)

       #
       n=sum(gammas.T)    # over Mixtures, zero order stat
       f=np.matrix(a) * np.matrix(gammas.T) # first order stat : [feat * GAMMAS']
       s=np.matrix(np.multiply(a, a)) * np.matrix(gammas.T)  #%2nd order stat (feat .* feat) * GAMMAS';

    return n, f, s, llk, n_frames

def GaussPosterior(a, mean1, cov_new, weights, n_mixtures, dim, n_frames):
    # Calculate posterior for a given data
    

    # denominator part of gaussian pdf
    B=(2*math.pi)
    B=math.pow(B,dim/2)
    
    A=B*(np.sqrt(np.prod(cov_new.T, axis=1)))
    A_a=weights/np.matrix(A.T) 


    #---
    B=(n_mixtures, n_frames)
    gammas = np.zeros(B, dtype=np.float64)

    

    for ii in range (0, n_mixtures):
            gammas[ii,:] = GaussFun(a, A_a[:,ii], mean1[:,ii], cov_new[:,ii],  n_frames, dim, n_mixtures)
    
    gamasum = sum(gammas)
    llk=np.log(gamasum)

    #normalize
    gammas=gammas/gamasum
    
    return gammas, llk


def GaussFun(a, aa, b, c, n_frames, dim, n_mixtures):
    #fit data to pdf
    auxC = -0.5/c
    aux = np.subtract(a.T, b.T)
    aux = np.power(aux, 2)         
    aux=aux.T
    Y = np.matrix(auxC.T) * aux    

     
    Y = np.exp(Y) * np.asscalar(aa)

    return Y

```

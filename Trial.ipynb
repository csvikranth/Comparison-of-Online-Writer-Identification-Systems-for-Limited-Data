```python
import xml.etree.ElementTree as ET  
```

```python
input_path='/content/drive/MyDrive/Datasets/query/11/11.inkml'
```

```python
tree = ET.parse(input_path)
root = tree.getroot()
root.tag
```

```python
for child in root:
  print(child.tag,child.attrib)
```

```python
for x in root[0]:
  print(x.text)
```

```python
l=[]
for i in range(1,len(root)):
  for x in root[i]:
    l.append(x.text)
    
```

```python
for y in l:
  if y=='\n\t\t\t':
    l.remove('\n\t\t\t')
print(len(l))
```

```python
b=[]
for x in l:
  k = list(map(str,x.split(',')))
  b=b+k
print(len(b))
```

```python
xy=[]
x_coordinates=[]
y_coordinates=[]
for a in b:
  a=a.strip()
  xy=list(map(str,a.split(" ")))
  xy=[int(z) for z in xy]
  x_coordinates.append(xy[0])
  y_coordinates.append(xy[1])
```

```python
import numpy as np
```

```python
print(len(x_coordinates))
time_stamp = np.arange(0,len(x_coordinates)*0.02,0.02)
print(time_stamp)
print(len(time_stamp))
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=100)
plt.plot(x_coordinates,y_coordinates)
```

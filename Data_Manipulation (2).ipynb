```python
import numpy as np
import csv
import os
import pickle
```

```python
features_path='/content/drive/MyDrive/Datasets/Split_8_2/Features_Train'
```

```python
writers=os.listdir(features_path)
```

```python
a=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
for writer in writers[:11]:
  print(writer)
  with open(os.path.join(features_path,writer)) as file:
    csvreader=csv.reader(file)
    c=0
    for row in csvreader:
      a[c].extend(row)
      c+=1
```

```python
from numpy import asarray
t = np.transpose(asarray(a))
print(t.shape)
```

```python
from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=25, init_params='kmeans', covariance_type='diag', max_iter=100)

gm.fit(t)
```

```python
mean = gm.means_
weights = gm.weights_
cov = gm.covariances_
```

```python
print(len(mean),len(weights),len(cov))
```
